```python
#create database
data = pd.read_csv ('IPO_Dashboard\IPO2.csv', encoding ='ISO-8859-1')
df = pd.DataFrame(data)

df= df.drop(columns=['Post_Money_Valuation', 'Pre_Money_Valuation', 'Initial_Shares', 'Initial_Filing_Range'])

df.head()
```

```python
#Clean Data
#Change NaN to 0

df= df.fillna(0)
df.head()
```

```python
#dropping unnecessary columns
Banking= df.drop(columns=['Company','Ticker','Shares \n','Amt_Raised\n'])
Banking=Banking.set_index('Pricing_Date')
Banking.head()
```

```python
#Dropping an empty line.
Banking=Banking.drop('3/24/2022')
Banking.head()
```

```python
Left_Bookrunner= Banking[['Left BR','Left_BR_%']]
Left_Bookrunner.head()
```

```python
Right_Bookrunner= Banking[['Right BR','Right_BR_%']]
Right_Bookrunner=Right_Bookrunner.replace(0,np.nan).dropna(axis=1,how="all")
Right_Bookrunner= Right_Bookrunner.dropna()
Right_Bookrunner.head()
```

```python
Third_Bookrunner= Banking[['Third BR','Third_BR_%']].replace(0,np.nan).dropna(axis=1,how="all")
Third_Bookrunner= Third_Bookrunner.dropna()
Third_Bookrunner.head()
```

```python
Fourth_Bookrunner= Banking[['Fourth BR','Fourth_BR_%']].replace(0,np.nan).dropna(axis=1,how="all")
Fourth_Bookrunner= Fourth_Bookrunner.dropna()
Fourth_Bookrunner.head()
```

```python
Left_CM= Banking[['Left CM','Left_CM_%']].replace(0,np.nan).dropna(axis=1,how="all")
Left_CM= Left_CM.dropna()
Left_CM.head()
```

```python
Right_CM= Banking[['Right CM','Right_CM_%']].replace(0,np.nan).dropna(axis=1,how="all")
Right_CM= Right_CM.dropna()
Right_CM.head(10)
```

```python
Third_CM= Banking[['Third CM','Third_CM_%']].replace(0,np.nan).dropna(axis=1,how="all")
Third_CM= Third_CM.dropna()
Third_CM.head(10)
```

```python
#Average Right Bookrunners
R_Bookrunner_AVG= Right_Bookrunner.groupby('Right BR').mean()*100
R_Bookrunner_AVG.head()
```

```python
#average percentage Left Bookrunners

L_Bookrunner_AVG= Left_Bookrunner.groupby('Left BR').mean()*100
L_Bookrunner_AVG.head()
```

```python
#average percentage of third BR
T_Bookrunner_AVG= Third_Bookrunner.groupby('Third BR').mean()*100
T_Bookrunner_AVG.head()
```

```python
#average percentage of fourth BR
F_Bookrunner_AVG= Fourth_Bookrunner.groupby('Fourth BR').mean()*100
F_Bookrunner_AVG.head()
```

```python
# average percentage left CM
L_CM_AVG= Left_CM.groupby('Left CM').mean()*100
L_CM_AVG.head()
```

```python
# average percentage right CM
R_CM_AVG= Right_CM.groupby('Right CM').mean()*100
R_CM_AVG.head()
```

```python
# average percentage third CM
T_CM_AVG= Third_CM.groupby('Third CM').mean()*100
T_CM_AVG.sort_values('Third CM', ascending=False)
T_CM_AVG.head()
```
